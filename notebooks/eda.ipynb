#### BASIC EDA

```python
# Plot histograms of each feature
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
sns.histplot(data=df, x='θ (°)', ax=axes[0])
axes[0].set_title('Inclination Histogram')
sns.histplot(data=df, x='Resg', ax=axes[1])
axes[1].set_title('Reynolds Gas Histogram')
sns.histplot(data=df, x='Resl', ax=axes[2])
axes[2].set_title('Reynolds Liquid Histogram')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
sns.histplot(data=df, x='θ (°)', hue="Regime", ax=axes[0])
axes[0].set_title('Inclination Histogram')
sns.histplot(data=df, x='Resg', hue="Regime", ax=axes[1])
axes[1].set_title('Reynolds Gas Histogram')
sns.histplot(data=df, x='Resl', hue="Regime", ax=axes[2])
axes[2].set_title('Reynolds Liquid Histogram')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
sns.scatterplot(data=df, x='θ (°)', y='αG (-)', ax=axes[0])
axes[0].set_title('Inclination Scatter Plot')
sns.scatterplot(data=df, x='Resg', y='αG (-)', ax=axes[1])
axes[1].set_title('Reynolds Gas Scatter Plot')
sns.scatterplot(data=df, x='Resl', y='αG (-)', ax=axes[2])
axes[2].set_title('Reynolds Liquid Scatter Plot')
plt.show()
```
